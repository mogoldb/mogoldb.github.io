This offseason in the National Football Leauge (NFL), two teams, the Chiefs and the Packers, who have two of the best receivers in the leauge, Tyreek Hill and Devante Adams, have traded those players away. In turn they got many draft picks and it seems as though these teams strategies are to have more good receivers rather than having 1-2 star receivers with a very below average the rest of the receiving corp. This brings up an interesting team building stragety question in a salary capped leauge: is it better to have a top heavy receiving corp with superstars that change the math of the game, or is it better to have a unit of more reliable receivers without one gamechanger?

First, I'll explain a little bit about the salary cap in the NFL. According to [Bleacher Report](https://bleacherreport.com/articles/1665623-how-does-the-salary-cap-work-in-the-nfl), "A salary cap is essentially an agreement between the league and players that places a limit on the amount of money a team can spend on salaries for players. The NFL uses a hard cap, meaning that no team is allowed to exceed the cap limit for any reason." The cap is what blances the leauge ensuring that stragtety has to be employed in team building rather than being able to buy the best team possible which can create a boring leauge with the richest teams dominating the game.

The salary cap forces teams to prioritize where they put their money. Building a team is like recruiting, you want the best players to come to you and one of the ways to do that is by paying them the most money. But you also want to make sure that you are not putting too much money in one area or player because if you do so, the salary cap will prevent you from being able to invest as much in the rest of the team. Two big things that need to be taken into account when thinking about money allocation in team buliding are positional value, and the value of each individual player. Pro Football Focus (PFF), a football analytics company, has a metric called Wins Above Replacement (WAR), which estimates roughly how many wins an individual player has contributed to his team. Meaning, you should subtract a players WAR from the team's total wins to say how many wins the team would have if this player were replaced with an average player. WAR is a way of calculating value of a specific player, but also tells us a lot about positional value. This chart from PFF shows the average WAR of the top 10 players from each position from 2019: 

<center><img src="pff_2019_war.png" width="600"></center>

Data shows that in the modern NFL, which is a passing leauge, after Quarterback, wide recievers are the most valuable position on the team. While there is a lot of data on individual player value and positional value, there isn't a lot of research on what combinations of players make for the best position group. As in, what combination of wide receivers give you the best receiving corp? On the one hand, receivers have the ability to completely change the math of the game espcially when it comes to star receivers. When one receiver needs to defenders on him at all times, or is so fast that the opposing team needs to adjust their entire defensive scheme for him, he's valuable and impacting the game even when he doesn't touch the ball. On the other hand, it's just one player, and if a great defensive coach can figure out how to shut him down and the rest of the receiving corp is not very reliable because all the money went into the star, is it better to have 2 or 3 above average players than being top heavy? Were the Cheifs and Packers right in opting to pay one individual player less and spreading there money across receivers more equally, or did the teams that aquired Tyreek Hill and Devante Adams have the right idea?

In this tutorial, we will inspect PFF data and come up with a model that can take in the performance grades of 4 receivers and predict predict the total receiving corp performance based on the performance. This can allow general managers (GMs) to look at the receivers they have and run the model with different free agent receivers and try to predict what the total performance of those hypothetical receiving corps will be.

## Part 1: Gathering the data

First, we need to gather and decide what data to use. While many people look at standard statistics in football, like touchdowns, percentage of targets caught, and yards per reception, a lot of these traditional statistics are flawed and don't provide context. Firstly, it only evaluates a players performance when they are targeted, so it doesn't account for how they impact the rest of the game. Additionally, football is the ultimate team sport and a lot of those traditional statistics are depenent on the performance of everyone around the receiver not just the receivers perofrmance. So for example, if a receiver compeltely runs the wrong route, but then the quarterback finds him anyways and he gets a touchdown the receiver gets credit for that when he actually did something wrong. What PFF does is evaluate every player on every play. This helps account for their perfomance when players are not touching the ball, in addition to isolating player performance so that the player does not get penalized for other players mistakes or credit for other players great plays. These grades are out of 100 and usually anything above 90 is considered elite but the average grade varies with position. PFF also has grades for team units, like grades for a teams offense, defense, receiving corp, secondary, etc. So we will use individual player grades and the team receiving corp grade to train our model.

One consideration we have to make is how many years of data do we want? On the one hand, more data helps us train our model better, but on the other hand the game of football has rapidly changed over the past 5-10 years so going to far back can train the model with data that is no longer relevant to today's NFL. In order decide how far to go back I wanted to find a point from one year to another that made it clear the NFL started valuing receivers signinficalty more. I looked at overthecap.com and looked at how much the teams that spend the most money on receivers spent. There was a big jump from the 2016 to 2017:

This influx started around the 2017 season, and 2017 feels like a good place to define a starting point for this newer era of football.

```python
recievers2021 = pd.read_csv("receiving_summary_2021.csv")
recievers2021['year'] = 2021 # add a column to represent year

recievers2020 = pd.read_csv("receiving_summary_2020.csv")
recievers2020['year'] = 2020

recievers2019 = pd.read_csv("receiving_summary_2019.csv")
recievers2019['year'] = 2019

recievers2018 = pd.read_csv("receiving_summary_2018.csv")
recievers2018['year'] = 2018

recievers2017 = pd.read_csv("receiving_summary_2017.csv")
recievers2017['year'] = 2017

# combine each year's data
receiver_summary = pd.concat([recievers2021, recievers2020, recievers2019, recievers2018, recievers2017])
receiver_summary = receiver_summary[(receiver_summary.position != 'HB')] # removing running backs
receiver_summary.head()
```

### Getting Team Reciver Unit Grades

For the team summaries PFF didn't allow me to download that data as a spreadsheet like I was able to for individual players recieving summaries.
```python
teams2021 = pd.read_csv("team_summary_2021.csv")
teams2021['year'] = 2021 # add a column to represent year

teams2020 = pd.read_csv("team_summary_2020.csv")
teams2020['year'] = 2020

teams2019 = pd.read_csv("team_summary_2019.csv")
teams2019['year'] = 2019

teams2018 = pd.read_csv("team_summary_2018.csv")
teams2018['year'] = 2018

teams2017 = pd.read_csv("team_summary_2017.csv")
teams2017['year'] = 2017

team_summaries = pd.concat([teams2021, teams2020, teams2019, teams2018, teams2017]) # combine each year's data
team_summaries.head()
```

## Part 2: Transforming the Data
Now we are not sure what sort of model will fit for this data, but we will start with a linear regression and run some tests to see if the data is linear. The regression will have the grades of every team's top 4 receivers as the independent variable and the teams total receiving grade will be the dependant variable. So we have to transform the top 4 receivers from each team on a given year into one row where there are columns for each receiver.

The first step is to add a row of team abbriviations to the team summaries table so we can match up teams and receivers from the receivers table. We will use the nfl_team.csv file for this.

```python
nfl_team_info = pd.read_csv("nfl_teams.csv") # NFL team data - name, team abbriviation, etc
nfl_team_info.head()
```

```python
# Rename the column to Name so we can merge with nfl_team_info to get Abbriviation
team_summaries = team_summaries.rename(columns={'TEAM': 'Name'}) 

rec_data = pd.merge(nfl_team_info, team_summaries, how ='inner', on=['Name'])

# rename columns so it'll match better with the receiver_summary naming conventions
rec_data = rec_data.rename(columns={'RECV': 'team_grades_pass_route', 'Abbreviation': 'team_name', 'Name': 'full_name'})

# drop data that is irrelevant
rec_data = rec_data.drop(columns=['ID', 'Conference', 'Division', 'PF', 'PA', 'OVER', 'OFF', 
                                  'PASS', 'PBLK', 'DEF', 'RDEF'])

rec_data.head()
```

```python
# drop irrelevant receiver data
receiver_grades = receiver_summary[['player', 'position', 'team_name', 'year', 'grades_pass_route', 'targets']]
receiver_grades.head()
```

Now that we have our team and receiver data formatted nicely we will merge the two datasets.

```python
merged = pd.merge(rec_data, receiver_grades, how ='outer', on=['year', 'team_name'])
merged.head()
```

But this merge isn't exactly what we wanted. We need each team and year combination to be one row which contains the team passing grade and the top 4 players on each team's individual grades. We need to do a bit more manipulation. We will go through each team and then each year for each of those teams and add a row to a new dataframe we are building.

Additionally, we are defining top 4 not by grade, but by number of targets. This way each player will be ordered in terms of how much they are relied on. So the receiver that gets the most targets is considered the number 1 receiver and not the receiver with the highest grade. If we sorted by the receiver with the highest grade, there could be receivers who have been on the field for 2 games and got 5 targets and have a really good grade. But that doesn't tell us much about the receiving corp and the players effect on the team.

```python
team_groupings = merged.groupby('team_name') # group by team name
team_names = nfl_team_info['Abbreviation'].tolist() # a list of all team names

df = pd.DataFrame()
i = 0
for name in team_names:
    # the data associated with the team we are currently evaluating and then group that data by years
    team = team_groupings.get_group(name).groupby('year') 
     # since some teams names have changed not all team names so only get the years associated with that team name
    years = list(team.groups.keys())
    for year in years:
        # get a dataframe with the top 4 receivers with the most targets in this team, year combination
        top4 = team.get_group(year).nlargest(n=4, columns=['targets'])
       
        # add the team data to a new df
        to_add = pd.DataFrame({'full_name': top4['full_name'].iloc[0], 
                               'team_name': name,
                               'team_grades_pass_route': top4['team_grades_pass_route'].iloc[0],
                               'year': year
                              }, index=[i])
        j = 1
        for index, row in top4.iterrows(): # go through each row of the top 4 players and add their grade as a column
            to_add['player{}_rec_grade'.format(j)] = row.grades_pass_route
            j += 1
        
        if df.empty:
            df = to_add
        else:
            df = pd.concat([df, to_add]) # in each year get the rows with the top 4 passing routes
        i += 1
df.head()
```

```python
print(df[pd.notnull(df["player1_rec_grade"])]["player1_rec_grade"].count())
print(df[pd.notnull(df["player2_rec_grade"])]["player2_rec_grade"].count())
print(df[pd.notnull(df["player3_rec_grade"])]["player3_rec_grade"].count())
print(df[pd.notnull(df["player4_rec_grade"])]["player4_rec_grade"].count())
df[df['player1_rec_grade'].isna()]
```

There are 161 rows of data and all the reception grades have 160 row entries. The only row that is missing player data is that last Okland Raiders one due to weirdness with their name changing in 2020 from OAK to LV because they moved to Los Vegas so we can just drop that column.

```python
df.dropna(inplace=True) # drop null column
```

## Part 3: Checking Linear Regression Assumptions

While we can run a linear regression on any dataset and we will get a model that can predict things, those predictions don't mean anything unless the data is linear. Meaning if the model doesn't fit the data, the model is useless in it's predictive powers. We must make sure that the 4 assumptions of a linear regression are being met before using a linear regression on the data. The 4 assumptions goes as follows:

1. A linear relationship exists
2. Normality of residuals
3. No multicollinearity among predictors
4. No autocorrelation of residuals
5. Constant Variance of residuals (this is known as 'Homoscedasticity')

We will use techniques discussed in [this](https://towardsdatascience.com/verifying-the-assumptions-of-linear-regression-in-python-and-r-f4cd2907d4c0) towards data science and [this](https://jeffmacaluso.github.io/post/LinearRegressionAssumptions/) article to evaluate if these assumptions are true.

First, lets establish linearity. To detect linearity we can inspect plots of observed vs. predicted values or residuals vs. predicted values. If the data is linear we would expect that the points are symmetrically distributed around a diagonal line in the observed vs. predicted values plot and a horizontal line for the residuals vs. predicted values plot. The variance should also be nearly constant.

We will start by using the data we formatted above to make a linear regression model:

```python
# independent variable is the receiver grades 
x = df[['player1_rec_grade', 'player2_rec_grade', 'player3_rec_grade', 'player4_rec_grade']]

# dependent variable is the team receiving corp grades 
y = df[['team_grades_pass_route']]

X_constant = sm.add_constant(x) # since intercept is not included by default we need to add a constant
lin_reg = sm.OLS(y, X_constant).fit()
print(lin_reg.summary())
```

Our r2 is pretty good, which can indicate linearity, but let use this linearity test provided in the towards data science article to look at the observed vs. predicted values and residuals vs. predicted values plots.

### 3.1. Testing Linearity

```python
sns.set_style('darkgrid')
sns.mpl.rcParams['figure.figsize'] = (15.0, 7.5)

def linearity_test(model, y):
    '''
    Function for visually inspecting the assumption of linearity in a linear regression model.
    It plots observed vs. predicted values and residuals vs. predicted values.
    
    Args:
    * model - fitted OLS model from statsmodels
    * y - observed values
    '''
    fitted_vals = model.predict()
    resids = model.resid

    fig, ax = plt.subplots(1,2)
    
    sns.regplot(x=fitted_vals, y=y, lowess=True, ax=ax[0], line_kws={'color': 'red'})
    ax[0].set_title('Observed vs. Predicted Values', fontsize=16)
    ax[0].set(xlabel='Predicted', ylabel='Observed')

    sns.regplot(x=fitted_vals, y=resids, lowess=True, ax=ax[1], line_kws={'color': 'red'})
    ax[1].set_title('Residuals vs. Predicted Values', fontsize=16)
    ax[1].set(xlabel='Predicted', ylabel='Residuals')
    
linearity_test(lin_reg, y)
```

We expected that if the data was linear for the first plot to have a diagonal line, the second to have a horizontal, and for the variance to be mostly constant. As you can see above, those conditions seem to be satisfied and therefore we can feel confident that this dataset satisfies the linearity assumption.

### 3.2. Testing Normaltiy of Residuals
There are several ways to evaluate the normality of the residuals. We can look QQ plots of the residuals. A QQ plot plots two sets of quantiles against each other if they both come from the same distribution, the points should form a roughly straight line. So lets inspect this qq plot to see if it looks like a straight line.

```python
sns.mpl.rcParams['figure.figsize'] = (8.0, 6.0)

sm.ProbPlot(lin_reg.resid).qqplot(line='s');
plt.title('Q-Q plot');
```

This line looks mostly straight with some outliers which tells us that the residuals have a normal distribution for the most part.

### 3.4. Testing no Autocorrelation of Residuals

Next, we need to there is autocorrelation in the residuals. The way we can test that is by preforming the Durbin-Watson to detect if there is a positive or negative correlation on the residuals. When the test produces values between 1.5 and 2.5 it means that there is no autocorrelation, when the results are between 0 and 2 it means there is positive autocorrelation, and when the result is from 2 to 4 there is a negative autocorrelation.

```python
print('Durbin-Watson:', durbin_watson(lin_reg.resid))
```

Since the value is between 1.5 and 2.5 we can say there is little to no autocorrelation and this assumption is satisfied.

### 3.5 Testing for Homoscedasticity of Residuals

Now, lets evaluate for homoscedasticity. There are several ways to do so. First, we can look at the residuals vs. predicted values from the linearity test and see if the variance seems to be uniform. For the most part, it did. But we will also run a test called the Breusch-Pagan test. In this test, the null hypothesis assumes homoscedasticity and a p-value below 0.05 will indicate that we need to reject the null hypothesis of homoscedasticity.

```python
def homoscedasticity_test(model):
    '''
    Function for testing the homoscedasticity of residuals in a linear regression model.
    It runs Breusch-Pagan and Goldfeld-Quandt tests.
    
    Args:
    * model - fitted OLS model from statsmodels
    '''
    fitted_vals = model.predict()
    resids = model.resid
    resids_standardized = model.get_influence().resid_studentized_internal

    bp_test = pd.DataFrame(sms.het_breuschpagan(resids, model.model.exog), 
                           columns=['value'],
                           index=['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value'])

    print('\n     ---- Breusch-Pagan test ----')
    print(bp_test)

homoscedasticity_test(lin_reg)
```

We have now gone throught the assumptions of a linear regression and satisfied them. Meaning we can now move forward and train our linear regression model. While we used an OLS regression above, we didn't split up our data into a testing and training split so we will do that and do the regression again.

## Part 4: Data Analysis

Previously, our model was trained on the entire dataset so the r2 value was based on values the model had already seen. That means we are evaluating the models accuracy based on data it trained on which is problematic because we want to see how it performs on data it's never seen before. To solve this issue, we will split the data up and train our model on most of the data but preserve a small portion of data to test the model on.

We are attempting to make a data model that can take in the top 4 most productive receivers on a team and predict the overall performance (the grade) of that receiving corp. In order to do this the model will be provided the receiver grades as the independent variable and the receiving unit grade will the the depenent variable.

When buliding the model initially, I had sorted players by grade in addition to only including grades and the model was clearly very off. There were also issues when changing the order of the grades put into the model because the model would way wr3 and wr4 more because the coefficients were higher. I wanted a way to normalize that a little more and though including targets would help. It did! I then realized that choosing the top 4 players and sorting by grade was a bad idea for the reasons I explained above so I started sorting by targets. Once I sorted by targets, I was able to remove targets as a feature the the model performed very slightly better. I decided not to include targets as a feature of the model, not because the model performed better that way, but because logically once the model is taking into account that the wr1 has the most targets wr2 has the second most and so on. Once the model has a way to weigh each receiver by their place on the depth chart the exact number of targets is not necessary. If you can get a model working with less features its better than adding more noise so I chose to leave the targets out of the training.

```python
x_train, x_test, y_train, y_test = ms.train_test_split(x, y, random_state=42)
print(x_train.size, x_test.size) # this preserves about 25% of the data for testing and 75% for the training
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_prediction = model.predict(x_test)

# predicting the accuracy score
score = met.r2_score(y_test, y_prediction)
print('r2 score is: ', score)
print('mean_sqrd_error is: ', met.mean_squared_error(y_test,y_prediction))
print('root_mean_squared error of is: ', np.sqrt(met.mean_squared_error(y_test,y_prediction)))
```

We now have a model that can be used to build the ultimate receiving corp. Now, since this model isn't exactly something we can look at lets pretend that we are GMs and it is the offseason.

```python
def print_prediction_against_exp(row):
    x = row[['player1_rec_grade', 'player2_rec_grade', 'player3_rec_grade', 'player4_rec_grade']]

    year = row['year'].values[0]
    pg1 = row['player1_rec_grade'].values[0]
    pg2 = row['player2_rec_grade'].values[0]
    pg3 = row['player3_rec_grade'].values[0]
    pg4 = row['player4_rec_grade'].values[0]
    
    prediction = model.predict(x)
    print('''Receiver corp grade prediction for:
            player 1 grade {},
            player 2 grade {},
            player 3 grade {},
            player 4 grade {},
            is: {}
            while the actual grade is: {}'''.format(pg1, pg2, pg3, pg4, 
                                                    prediction[0], row['team_grades_pass_route'].values[0]))

def print_prediction(grades):
    grade_lst = []
    
    n = len(grades)
    for _ in n:
        temp = max(grades, key=lambda k: grades[k])
        grade_lst.append(temp)
        del grades[temp]
        
    print(grade_lst)
    prediction = model.predict([grade_lst])

    print('''Receiver corp grade prediction for:
            player 1 grade {},
            player 2 grade {},
            player 3 grade {},
            player 4 grade {},
            is: {}'''.format(pg1, pg2, pg3, pg4, prediction[0]))

print_prediction(87.2, 69.6, 64.6, 70.4)
print_prediction(87.2, 85.6, 70.4, 69)

print_prediction_against_exp(df[(df.team_name == 'KC') & (df.year == 2021)])
print_prediction_against_exp(df[(df.team_name == 'KC') & (df.year == 2020)])
print_prediction_against_exp(df[(df.team_name == 'SF') & (df.year == 2021)])
print_prediction_against_exp(df[(df.team_name == 'TB') & (df.year == 2021)])
print_prediction_against_exp(df[(df.team_name == 'TB') & (df.year == 2020)])
```
